```python
from enum import Enum
from pydantic import BaseModel, Field, model_validator
from typing import List, Optional
import re

class BusinessObjectiveEnum(Enum):
    SALES_INCREASE = 'Sales Increase'
    COST_REDUCTION = 'Cost Reduction'
    MARKET_EXPANSION = 'Market Expansion'
    PRODUCT_DEVELOPMENT = 'Product Development'
    CUSTOMER_SATISFACTION = 'Customer Satisfaction'
    OTHER = 'Other'

class BusinessObjective(BaseModel):
    category: BusinessObjectiveEnum


class Step(BaseModel):
    step_id: int
    description: str
    rationale: str
    expected_outcomes: str
    metrics_for_success: str

class ActionPlan(BaseModel):
    steps: List[Step]

class Solution(BaseModel):
    solution_id: int
    description: str
    rationale: str
    action_plan: ActionPlan

class BusinessProblemModel(BaseModel):
    business_problem: str
    solutions: List[Solution]
    business_objective: BusinessObjective

    # After the model has been generated, let's check if the solutions contain keywords from the business problem
    # If they do, we'll throw an error, return the filtered solutions and the error information, then have the llm try again.
    @model_validator(mode="after")
    def filter_solutions(self):
        # Extract keywords from the business problem
        keywords = re.findall(r'\w+', self.business_problem.lower())

        # Filter out solutions that contain keywords from the business problem
        tried_solutions = [
            solution for solution in self.solutions
            if any(keyword in solution.description.lower() for keyword in keywords)
        ]
        if len(tried_solutions) > 0:
            raise ValueError(f"Filtered out solutions that contain keywords from the business problem: {tried_solutions}")
        return self

```
